```python
import tensorflow as tf
from tensorflow.keras.applications import InceptionV3
from tensorflow.keras.applications.inception_v3 import preprocess_input
from tensorflow.keras.preprocessing import image
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
import numpy as np
```

```python
base_model = InceptionV3(weights='imagenet', include_top=False)
```

```python
train_dir = r"D:/download/archive/chest_xray/chest_xray/train"
test_dir = r"D:/download/archive/chest_xray/chest_xray/test"
```

```python
import os
import numpy as np
from PIL import Image, ImageOps
import tensorflow as tf
from tensorflow.keras.applications.inception_v3 import InceptionV3, preprocess_input
def process_image(img_path):
    try:
        img = Image.open(img_path)
        img = img.resize((150, 150))  # InceptionV3 模型的输入大小
        img = ImageOps.colorize(img, white="white", black="black")
        img_array = np.array(img)
        img_array = preprocess_input(img_array)  # 预处理图像，使其适用于 InceptionV3 模型
        img_array = np.expand_dims(img_array, axis=0)  # 添加 batch 维度
        features = base_model.predict(img_array)
        return features.flatten()
    except Exception as e:
        print("Error processing image:", e)
        return None
```

```python
X_train = []
Y_train = []
err_counter = 1

# 遍历训练集文件夹
for class_label in os.listdir(train_dir):
    print(class_label)
    class_path = os.path.join(train_dir, class_label)
    
    # 遍历每个类别的图像
    for img_name in os.listdir(class_path):
        img_path = os.path.join(class_path, img_name)
        
        # 处理图像并提取特征
        features = process_image(img_path)
        
        if features is not None:
            X_train.append(features)
            Y_train.append(class_label)

# 将列表转换为 numpy 数组
X_train = np.array(X_train)
Y_train = np.array(Y_train)
```

```python
X_test = []
Y_test = []
err_counter = 1

# 遍历训练集文件夹
for class_label in os.listdir(test_dir):
    print(class_label)
    class_path = os.path.join(test_dir, class_label)
    
    # 遍历每个类别的图像
    for img_name in os.listdir(class_path):
        img_path = os.path.join(class_path, img_name)
        
        # 处理图像并提取特征
        features = process_image(img_path)
        
        if features is not None:
            X_test.append(features)
            Y_test.append(class_label)

# 将列表转换为 numpy 数组
X_test = np.array(X_test)
Y_test = np.array(Y_test)
```

```python
from sklearn.preprocessing import LabelEncoder

# 使用LabelEncoder将字符串标签转换为二进制形式
le = LabelEncoder()
Y_train_binary = le.fit_transform(Y_train)
Y_test_binary = le.transform(Y_test)

# 训练 Logistic Regression 模型
lr_model = LogisticRegression()
lr_model.fit(X_train, Y_train_binary)

# 训练 Random Forest 模型
rf_model = RandomForestClassifier()
rf_model.fit(X_train, Y_train_binary)

# 预测概率
lr_probs = lr_model.predict_proba(X_test)[:, 1]
rf_probs = rf_model.predict_proba(X_test)[:, 1]

# 计算 ROC 曲线和 AUC
fpr_lr, tpr_lr, _ = roc_curve(Y_test_binary, lr_probs)
fpr_rf, tpr_rf, _ = roc_curve(Y_test_binary, rf_probs)

roc_auc_lr = auc(fpr_lr, tpr_lr)
roc_auc_rf = auc(fpr_rf, tpr_rf)

# 画 ROC 曲线
plt.figure(figsize=(8, 6))
plt.plot(fpr_lr, tpr_lr, label=f'Logistic Regression (AUC = {roc_auc_lr:.2f})', color='darkorange')
plt.plot(fpr_rf, tpr_rf, label=f'Random Forest (AUC = {roc_auc_rf:.2f})', color='green')
# 绘制对角线
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')

plt.legend(loc='lower right')
plt.show()
```
